# github에 저장한 비디오에 대한 cv.VideoCapture가 실패하는 이유

- 설명은 다음 글을 참고할 것.
https://blog.naver.com/secutron/223121744063

```python
import cv2
from google.colab.patches import cv2_imshow

url ='https://github.com/secutron/TesTime/raw/main/data/sample_cctv.mp4'
cap = cv2.VideoCapture(url)

# 비디오의 첫 번째 영상만 디스플레이
if cap.isOpened():
    is_success, frame = cap.read()
    if is_success:
        cv2_imshow(frame)
else:
    print('에러!')

cap.release()
```
